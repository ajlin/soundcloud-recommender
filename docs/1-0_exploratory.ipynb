```python
import requests
import time


def collect(start_url):
    params = {'client_id':'1dff55bf515582dc759594dac5ba46e9','order':'created_at', 'limit':100,'linked_partitioning':1}
    pages=100
    post_ = []
    count_ = 1
    url = 'https://api.soundcloud.com/users/743108/favorites/'
    for i in range(pages):
        if url == None:
            print('count: ',count_)
            break
        else:
            res = requests.get(url,params=params)
            res.raise_for_status()
            if res.status_code == 200:
                post_.extend(res.json()['collection'])
                print(count_)
                count_+=1
                try:
                    url = res.json()['next_href']
                except KeyError:
                    url = None
            else:
                print(res.status_code)
                break
            time.sleep(1)
    return pd.DataFrame()
```

```python
params = {'client_id':'1dff55bf515582dc759594dac5ba46e9','order':'created_at', 'limit':100,'linked_partitioning':1}

def collect(category_id, category='users', subcategory='favorites',params=params):
    url = 'https://api.soundcloud.com/'
    url += category
    url += '/'
    url += category_id
    url += '/'
    url += subcategory
    url += '/'
    url = str(url)
    print(url)
    limit = 100
    post_ = []
    count_ = 0
    for i in range(pages):
        if url == None:
            print('done: ',count_)
            break
        else:
            res = requests.get(url,params=params)
            res.raise_for_status()
            if res.status_code == 200:
                post_.extend(res.json()['collection'])
                count_+= len(res.json()['collection'])
                print(count_)
                try:
                    url = res.json()['next_href']
                except KeyError: #if no next_href
                    url = None
            else:
                print('error.  status code: ',res.status_code)
                break
            time.sleep(1)
    return pd.DataFrame(post_)

```
